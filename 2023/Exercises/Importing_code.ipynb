You can import code from existing python files. Here, we will import functions from the *functions.py* file. Looks at the animation below how the .py files are generated in colab.

To import functions for a python file, we need to location and the definitions we want to import.

```
from <which file> import <which definitions>
```

````python
from functions import add
````

````python
add(1,2)
````

````python
multiple(1,2)
````

We have not imported function *multiple*. So, we get an error. We can do this in couple ways. First,

````python
from functions import add, multiple

multiple(1,2)
````

Else, we could all definitions from the functions.py with wildcard *

````python
from functions import *

multiple(1,2)
````

We can also use the dotted notation. Here, we import the *functions.py* file. Then access the functions with the dot (.) notation.

````python
import functions

functions.add(2,3)
````
